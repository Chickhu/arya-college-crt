MATPLOTLIB SACTTER

```python
x=np.array([1,2,4,5,7,8,9,4,7,13])
y=np.array([1,5,6,7,8,13,14,15,6,8])   
plt.scatter(x,y)
plt.show()
```

```python
x1=np.array([1,2,4,5,7,8,9,4,7,13])   #comparison between two scatter
y1=np.array([1,5,6,7,8,13,14,15,6,8])   
x2=np.array([1,9,4,8,7,8,9,0,9,13])
y2=np.array([1,3,6,7,8,1,14,1,5,8])  
plt.scatter(x1,y1) 
plt.scatter(x2,y2)
plt.show()
```

```python
x1=np.array([1,2,4,5,6,7,8])   #comparison between two scatter
y1=np.array([1,5,6,7,4,5,6]) 
c=np.array(['red','green','blue','yellow','pink','black','brown'])   #c is used to give different colour to each lable
plt.scatter(x1,y1,c=c) 
plt.show()
```

```python
x1=np.array([1,2,4,5,6,7,8])   #comparison between two scatter
y1=np.array([1,5,6,7,4,5,6]) 
size=np.array([10,20,30,40,50,8,9])   #s is used to give different size
plt.scatter(x1,y1,c=size) 
plt.show()
```

```python
x1=np.array([1,2,4,5,6,7,8])   #comparison between two scatter
y1=np.array([1,5,6,7,4,5,6]) 
size=np.array([10,20,30,40,50,8,9])   #s is used to give different size
plt.scatter(x1,y1,s=size,alpha=0.5) 
plt.show()
```

```python
x1=np.array([1,2,4,5,6,7,8])   #comparison between two scatter
y1=np.array([1,2,4,5,6,7,8]) 
size=np.array([10,20,30,40,50,8,9])   #s is used to give different size

c=np.array(['red','green','blue','yellow','pink','black','brown'])
plt.scatter(x1,y1,s=size,c=c,alpha=1)   #alpha used for intensity bw 0 and 1
plt.show()
```

```python
x1=np.array([1,2,4,5,6,7,8])   #comparison between two scatter
y1=np.array([1,2,4,5,6,7,8]) 
size=np.array([10,20,30,40,50,8,9])   #s is used to give different size

c=np.array(['red','green','blue','yellow','pink','black','brown'])
plt.scatter(x1,y1,s=size,c=c,alpha=0.5,cmap="nipy_spectral") 
#cmap=virdis,nipy_spectral
plt.colorbar()   #colorbar used to show colorline outside the graph
plt.show()
```

```python
y=np.array([1,2,3,4])   # to draw piechart
plt.pie(y)
plt.show()
```

```python
y=np.array([1,2,3,4])
mylable=['oxygen','co2','n2','no2']   #mylable gives different label of each data
plt.pie(y,labels=mylable)
plt.show()
```

```python
y=np.array([1,2,3,4])
mylable=['oxygen','co2','n2','no2']
plt.pie(y,labels=mylable,startangle=90)   #startangle used to shift 90 degree(staring)
plt.show()
```

```python
y=np.array([1,2,3,4])
mylable=['oxygen','co2','n2','no2']
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylable,startangle=90,explode=myexplode)    #exclude is used to move out a part from pirchart
plt.show()
```

```python
y=np.array([1,2,3,4])
mylable=['oxygen','co2','n2','no2']
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylable,startangle=90,explode=myexplode,shadow=True)  #shadow for 2d print
plt.show()
```

```python
y=np.array([1,2,3,4])
mylable=['oxygen','co2','n2','no2']
myexplode=[0.2,0,0,0]
mycolor=['red','orange','blue','yellow']     #autopct used to divide in percentagr ratio
plt.pie(y,labels=mylable,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.show()
```

```python
y=np.array([1,2,3,4])
mylable=['oxygen','co2','n2','no2']
myexplode=[0.2,0,0,0]  
mycolor=['red','orange','blue','yellow']    #legend used to show color outside piechart
plt.pie(y,labels=mylable,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.legend(title='color')
plt.show()
```
